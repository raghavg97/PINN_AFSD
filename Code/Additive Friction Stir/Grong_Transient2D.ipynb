```python
loss_thresh = 25000
level = "_low"
label = "Burgers_stan" + level

T0 = 300
pi = 3.14
rho = 1
c_p = 1
k = 0.001

x = np.linspace(-1,1,200).reshape(-1,1)
y = np.linspace(-1,1,200).reshape(-1,1)
t = np.linspace(0,10,200).reshape(-1,1)

X,Y,T = np.meshgrid(x,y,t)

X = X.flatten('F').reshape(-1,1)
Y = Y.flatten('F').reshape(-1,1)
T = T.flatten('F').reshape(-1,1)
  
xyt = np.hstack((X,Y,T))

# bound_pts_1 = (X == 0).reshape(-1,)
# bound_pts_2 = np.logical_and(Y == 0,X != 0).reshape(-1,)
# bound_pts_3 = np.logical_and(X == 1,Y != 0).reshape(-1,) 
# bound_pts_4 = np.logical_and(Y == 1,X != 1).reshape(-1,) 

# xy_bound_1 = xy[bound_pts_1,:]
# xy_bound_2 = xy[bound_pts_2,:]
# xy_bound_3 = xy[bound_pts_3,:]
# xy_bound_4 = xy[bound_pts_4,:]

# u_bound_1 = 1000*np.ones((np.shape(xy_bound_1)[0],1))
# u_bound_2 = 800*np.ones((np.shape(xy_bound_2)[0],1))
# u_bound_3 = 500*np.ones((np.shape(xy_bound_3)[0],1))
# u_bound_4 = np.zeros((np.shape(xy_bound_4)[0],1))

# xy_bound = np.vstack((xy_bound_1,xy_bound_2,xy_bound_3,xy_bound_4))
# u_bound = np.vstack((u_bound_1,u_bound_2,u_bound_3,u_bound_4))

# xy_test_tensor = torch.from_numpy(xy).float().to(device)

lb_xyt = xyt[0]
ub_xyt = xyt[-1]

r = np.linspace(0,1,200).reshape(-1,1)
theta = np.linspace(0,2*np.pi,200).reshape(-1,1)
t = np.linspace(0,10,200).reshape(-1,1)

R,Theta,T = np.meshgrid(r,theta,t)

R = R.flatten('F').reshape(-1,1)
Theta = Theta.flatten('F').reshape(-1,1)
T = T.flatten('F').reshape(-1,1)
  
rthetat = np.hstack((R,Theta,T))

lb_rthetat = rthetat[0]
ub_rthetat = rthetat[-1]
```

```python
def trainingdata(N_T,N_f,seed):
    '''Boundary Conditions''' 
    
    np.random.seed(seed)
    
    #Side 1 
    x_1 = x[0]*np.ones((N_T,1))
    y_1 = np.random.uniform(y[0],y[-1],(N_T,1))
    t_1 = np.random.uniform(t[0],t[-1],(N_T,1))
    xyt_1 = np.hstack((x_1,y_1,t_1))
    u_1 = T0*np.ones((N_T,1))
    
    #Side 2
    x_2 = np.random.uniform(x[0],x[-1],(N_T,1))
    y_2 = y[0]*np.ones((N_T,1))
    t_2 = np.random.uniform(t[0],t[-1],(N_T,1))
    xyt_2 = np.hstack((x_2,y_2,t_2))
    u_2 = T0*np.ones((N_T,1))
    
    #Side 3
    x_3 = x[-1]*np.ones((N_T,1))
    y_3 = np.random.uniform(y[0],y[-1],(N_T,1))
    t_3 = np.random.uniform(t[0],t[-1],(N_T,1))
    xyt_3 = np.hstack((x_3,y_3,t_3))
    u_3 = T0*np.ones((N_T,1))
    
    #Side 4
    x_4 = np.random.uniform(x[0],x[-1],(N_T,1))
    y_4 = y[-1]*np.ones((N_T,1))
    t_4 = np.random.uniform(t[0],t[-1],(N_T,1))
    xyt_4 = np.hstack((x_4,y_4,t_4))
    u_4 = T0*np.ones((N_T,1))
    
    
    xy_0 = np.random.uniform(x[0],x[-1],(N_T,2))
    t_0 = t[0]*np.ones((N_T,1))
    xyt_0 = np.hstack((xy_0,t_0))
    u_0 = T0*np.ones((N_T,1))
    
    xyt_BC = np.vstack((xyt_1,xyt_2,xyt_3,xyt_4,xyt_0)) #choose indices from  set 'idx' (x,t)
    u_BC = np.vstack((u_1,u_2,u_3,u_4,u_0))
    
    '''Collocation Points'''

    # Latin Hypercube sampling for collocation points 
    # N_f sets of tuples(x,t)
    x01 = np.array([[0.0,1.0],[0.0,1.0],[0.0,1.0]])
    sampling = LHS(xlimits=x01,random_state =seed)
    samples = sampling(N_f)
    
    xyt_coll = lb_xyt + (ub_xyt - lb_xyt)*samples
    
    rthetat_coll = lb_rthetat + (ub_rthetat - lb_rthetat)*samples
    x_coll2 = rthetat_coll[:,0]*np.sin(rthetat_coll[:,1])
    y_coll2 = rthetat_coll[:,0]*np.cos(rthetat_coll[:,1])
    xyt_coll2 = np.hstack((x_coll2.reshape(-1,1),y_coll2.reshape(-1,1),rthetat_coll[:,2].reshape(-1,1)))
    
    
    #Additional Collocation 
    # x_c_coll = np.random.normal(0,0.05,(N_T,1))
    # t_c_coll = np.abs(np.random.normal(0,0.05,(N_T,1)))
    # xt_c_coll = np.hstack((x_c_coll,t_c_coll))

    xyt_coll = np.vstack((xyt_coll, xyt_coll2, xyt_BC)) # append training points to collocation points 

    return xyt_coll, xyt_BC, u_BC
```

```python
class Sequentialmodel(nn.Module):
    
    def __init__(self,layers):
        super().__init__() #call __init__ from parent class 
              
        'activation function'
        self.activation = nn.Tanh()

     
        'loss function'
        self.loss_function = nn.MSELoss(reduction ='mean')
        
        'Initialise neural network as a list using nn.Modulelist'  
        self.linears = nn.ModuleList([nn.Linear(layers[i], layers[i+1]) for i in range(len(layers)-1)])
        self.iter = 0
        
        # std = gain * sqrt(2/(input_dim+output_dim))
        for i in range(len(layers)-1):
            nn.init.xavier_normal_(self.linears[i].weight.data, gain=1.0)
            # set biases to zero
            nn.init.zeros_(self.linears[i].bias.data)   
        
        beta_mean = 1.0*torch.ones((50,len(layers)-2))
        beta_std = 0.1*torch.ones((50,len(layers)-2))
        
        self.beta = Parameter(torch.normal(beta_mean,beta_std))
        self.beta.requiresGrad = True
            
    'foward pass'
    def forward(self,xyt):
        if torch.is_tensor(xyt) != True:         
            xyt = torch.from_numpy(xyt)                
        
        ubxyt = torch.from_numpy(ub_xyt).float().to(device)
        lbxyt = torch.from_numpy(lb_xyt).float().to(device)
    
                      
        #preprocessing input 
        xyt = 2.0*(xyt - lbxyt)/(ubxyt - lbxyt) - 1.0
        
        #convert to float
        a = xyt.float()
        
        for i in range(len(layers)-2):
            z = self.linears[i](a)
            z1 =self.activation(z)
            a = z1 + self.beta[:,i]*z*z1
       
            
        a = self.linears[-1](a) 
         
        return a
                        
    def loss_BC(self,xyt,u):
                
        loss_bc = self.loss_function(self.forward(xyt), u)
                
        return loss_bc
    
    def loss_PDE(self, xyt_coll, f_hat):
        
        g = xyt_coll.clone()             
        g.requires_grad = True
        u = self.forward(g) 
        
        u_x_y_t = autograd.grad(u,g,torch.ones([xyt_coll.shape[0], 1]).to(device), retain_graph=True, create_graph=True,allow_unused = True)[0]
        
        u_xx_yy_tt = autograd.grad(u_x_y_t,g,torch.ones(xyt_coll.shape).to(device), create_graph=True,allow_unused = True)[0]

        du_dt = u_x_y_t[:,[2]]
        
        d2u_dx2 = u_xx_yy_tt[:,[0]]
        d2u_dy2 = u_xx_yy_tt[:,[1]]
        
        #Heat source Gaussian Distributed in x and t
        sigma_x = 0.05
        sigma_t = 0.1
        exp_term1 = (torch.square(g[:,0].reshape(-1,1)) + torch.square(g[:,1].reshape(-1,1)))/sigma_x**2
        exp_term2 =  torch.square(g[:,2].reshape(-1,1))/sigma_t**2
        
        Q = 100.0*torch.exp(-0.5*(exp_term1 + exp_term2))/(np.sqrt(((2*pi)**3)*sigma_x*sigma_x*sigma_t)) 
        
        f = rho*c_p*du_dt - k*(d2u_dx2+d2u_dy2) - Q
        
        loss_f = self.loss_function(f,f_hat)
                
        return loss_f
    
    def loss(self,xyt_BC,u_BC,xyt_coll,f_hat):

        loss_BC = self.loss_BC(xyt_BC,u_BC)
        loss_f = self.loss_PDE(xyt_coll,f_hat)
        
        loss_val = loss_BC + loss_f
        
        return loss_val
         
    'test neural network'
    def test(self):
        # u_pred = self.forward(xt_test_tensor)
        # u_pred = u_pred.cpu().detach().numpy()
        u_pred = 0
   
        return u_pred

    def test_loss(self):
        u_pred = self.test()
        
        # test_mse = np.mean(np.square(u_pred.reshape(-1,1) - u_true.reshape(-1,1)))
        # test_re = np.linalg.norm(u_pred.reshape(-1,1) - u_true.reshape(-1,1),2)/u_true_norm
        test_mse = 0
        test_re = 0
        
        return test_mse, test_re 
```

```python
def data_update(loss_np):
    train_loss.append(loss_np)

    
    test_mse, test_re = PINN.test_loss()
    test_mse_loss.append(test_mse)
    test_re_loss.append(test_re)
```

```python
def train_step(xyt_BC,u_BC,xyt_coll,f_hat,seed):

    def closure():
        optimizer.zero_grad()
        loss = PINN.loss(xyt_BC,u_BC,xyt_coll,f_hat)
        loss.backward()
        #print(loss.cpu().detach().numpy())
        return loss

    optimizer.step(closure)
```

```python
def train_model(max_iter,rep): 
    print(rep) 
    torch.manual_seed(rep*9)
    start_time = time.time() 
    thresh_flag = 0

    xyt_coll_np_array, xyt_BC_np_array, u_BC_np_array = trainingdata(N_T,N_f,rep*22)
        
    xyt_coll = torch.from_numpy(xyt_coll_np_array).float().to(device)
    xyt_BC = torch.from_numpy(xyt_BC_np_array).float().to(device)
    u_BC = torch.from_numpy(u_BC_np_array).float().to(device)
        
    f_hat = torch.zeros(xyt_coll.shape[0],1).to(device)
    

    for i in range(max_iter):
        train_step(xyt_BC,u_BC,xyt_coll,f_hat,i)
        loss_np = PINN.loss(xyt_BC,u_BC,xyt_coll,f_hat).cpu().detach().numpy()
        
        if(thresh_flag == 0):
            if(loss_np < loss_thresh):
                time_threshold[rep] = time.time() - start_time
                epoch_threshold[rep] = i+1          
                thresh_flag = 1       
        data_update(loss_np)
        
        print(i,"Train Loss",train_loss[-1],"Test MSE",test_mse_loss[-1],"Test RE",test_re_loss[-1])   
        
    elapsed_time[rep] = time.time() - start_time  
    print('Training time: %.2f' % (elapsed_time[rep]))
```
