```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define the parameters
L = 1.0
N = 50
dx = L / N
x = np.linspace(0, L, N+1)

# Define the source term f(x)
f = np.ones(N+1)  # Example: f(x) = 1 for simplicity

# Define the boundary conditions
u0 = 0.0  # u(0) = u0
uL = 0.0  # u(L) = uL

# Set up the matrix A and vector b
A = np.zeros((N+1, N+1))
b = np.zeros(N+1)

# Fill the matrix A
for i in range(1, N):
    A[i, i-1] = 1.0 / dx**2
    A[i, i] = -2.0 / dx**2
    A[i, i+1] = 1.0 / dx**2
    b[i] = f[i]

# Apply boundary conditions
A[0, 0] = 1.0
b[0] = u0
A[N, N] = 1.0
b[N] = uL

# Solve the linear system
u = np.linalg.solve(A, b)

# Print the solution
print("Solution u:", u)
```

```python
plt.plot(x,u)
```
