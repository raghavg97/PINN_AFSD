```python
fig,ax = plt.subplots()
ax.plot(x,T,'r-', label = 'predicted')
ax.plot(x,(2.5*np.square(x) + 3),'k',label = 'true')
```

```python
#ChatGpt Code
x_start = 0
x_end = 1
num_intervals = 5
delta_x = (x_end - x_start) / num_intervals

# Discretize the domain
x = np.linspace(x_start, x_end, num_intervals + 1)

# Initialize T array
T = np.zeros(num_intervals + 1)

# Boundary condition
T[0] = 3

# Finite difference method
for i in range(num_intervals):
    T[i + 1] = T[i] + 5 * x[i] * delta_x

# Output results
```

```python
fig,ax = plt.subplots()
ax.plot(x,T,'r-', label = 'predicted')
ax.plot(x,(2.5*np.square(x) + 3),'k',label = 'true')
```
